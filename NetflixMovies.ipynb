```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
movie['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#1f77b4', '#ff7f0e'])
plt.title('Distribution of Movies vs TV Shows on Netflix')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
movie['country'].value_counts().head(10).plot(kind='bar', color='darkgreen')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
movie['release_year'].value_counts().sort_index().plot(kind='line', marker='o', color='red')
plt.title('Netflix Content Release Trend')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid()
plt.show()
```

```python
from wordcloud import WordCloud

text = ' '.join(movie['genres'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='black', colormap='Set2').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Popular Genres on Netflix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
print(movie.columns)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=movie['popularity'], y=movie['rating'], alpha=0.6, color='darkviolet')
plt.title('popularity vs rating of Netflix Content')
plt.xlabel('popularity')
plt.ylabel('rating')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_columns = ['popularity', 'rating', 'vote_average', 'vote_count', 'release_year']
sns.pairplot(movie[numeric_columns], diag_kind='kde')
plt.suptitle('Pairplot of Netflix Movie Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
movie['rating'].value_counts().plot(kind='bar', color='darkorange')
plt.title('Distribution of Movie Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
movie.groupby('release_year')['vote_average'].mean().plot(kind='line', marker='o', color='purple')
plt.title('Average Vote Score Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Average Vote Score')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
movie['language'].value_counts().head(10).plot(kind='bar', color='brown')
plt.title('Most Common Languages on Netflix')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_columns = movie.select_dtypes(include=['number']).columns
correlation_matrix = movie[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Netflix Movie Features')
plt.show()
```
